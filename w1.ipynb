```python
def manager_sale_explorer():
    c1 = np.random.choice(df_36_months['CONTACT_ID'].values)
    sample_manager = df_36_months.loc[df_36_months['CONTACT_ID'] == c1]
    plt.plot(sample_manager['refresh_date'], sample_manager['no_of_sales_12M_1'])
```
